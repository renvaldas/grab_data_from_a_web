```python
import pandas as pd
#tables = pd.read_html("https://apps.sandiego.gov/sdfiredispatch/")
url = "http://yann.lecun.com/exdb/mnist/"
tables = pd.read_html(url)
#print(tables[0].head())
tables[0].head(10)
```

```python
# letâ€™s tell Pandas that row 0 of the table has column headers and ask it to convert text-based dates into time objects
# calls_df, = pd.read_html("http://apps.sandiego.gov/sdfiredispatch/", header=0, parse_dates=["Call Date"])
calls_df, = pd.read_html(url, header=0)
calls_df.head()
```

```python
# Wish the data was available as json records
# calls_df, = pd.read_html("http://apps.sandiego.gov/sdfiredispatch/", header=0, parse_dates=["Call Date"])
# print(calls_df.to_json(orient="records", date_format="iso"))
calls_df, = pd.read_html(url, header=0)
print(calls_df[:5].to_json(orient="records"))
```

```python
# save the data right to a CSV or XLS file
calls_df.to_csv("calls.csv", index=False)
```
